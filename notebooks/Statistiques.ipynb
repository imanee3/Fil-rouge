## 1. Analyse statistique

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.cluster import KMeans

# ============================================================
# 1) LOAD DATA
# ============================================================
df = pd.read_csv(r"C:\Users\imane\Desktop\Fil-rouge\data\gold\f1_final_enriched.csv")

print("Shape:", df.shape)
print(df.head())

# garder seulement la course pour la plupart des analyses métier
df_r = df[df["session"] == "R"].copy()

print("\nShape Race only:", df_r.shape)

# ============================================================
# 2) ROI PILOTE
# ============================================================
roi_driver = (
    df_r.groupby("driverName")["driverROI"]
    .mean()
    .sort_values()
)

print("\n==============================")
print("ROI PILOTE")
print("==============================")
print("\nTop pilotes les plus rentables (ROI faible) :")
print(roi_driver.head(10))

print("\nPilotes les moins rentables (ROI élevé) :")
print(roi_driver.tail(10))

plt.figure(figsize=(10, 5))
roi_driver.head(10).sort_values(ascending=False).plot(kind="barh")
plt.title("Top 10 ROI Pilote (plus rentable = plus faible)")
plt.xlabel("Driver ROI")
plt.ylabel("Driver")
plt.show()

# ============================================================
# 3) BUDGET VS PERFORMANCE
# ============================================================
team_budget_perf = (
    df_r.groupby("constructorName")[["costCapUSD", "constructorPoints", "budgetEfficiency", "pointsPerMillionUSD"]]
    .mean()
    .sort_values("constructorPoints", ascending=False)
)

print("\n==============================")
print("BUDGET VS PERFORMANCE")
print("==============================")
print(team_budget_perf)

plt.figure(figsize=(8, 5))
sns.regplot(data=df_r, x="costCapUSD", y="constructorPoints")
plt.title("Budget vs Performance")
plt.xlabel("Cost Cap USD")
plt.ylabel("Constructor Points")
plt.show()

print("\nCorrélation budget vs points :")
print(df_r[["costCapUSD", "constructorPoints"]].corr())

# ============================================================
# 4) CORRELATION MATRIX
# ============================================================
corr_cols = [
    "driverPoints",
    "raceGain",
    "driverROI",
    "driverEfficiency",
    "budgetEfficiency",
    "pointsPerMillionUSD",
    "speedEfficiency",
    "paceAdvantage",
    "driverPerformanceScore"
]

corr = df_r[corr_cols].corr()

print("\n==============================")
print("CORRELATION MATRIX")
print("==============================")
print(corr)

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix KPI")
plt.show()

# ============================================================
# 5) TESTS STATISTIQUES (PEARSON)
# ============================================================
print("\n==============================")
print("TESTS STATISTIQUES (PEARSON)")
print("==============================")

# 5.1 ROI vs Performance
subset_roi = df_r.dropna(subset=["driverROI", "driverPerformanceScore"])
corr_roi, p_roi = pearsonr(subset_roi["driverROI"], subset_roi["driverPerformanceScore"])

print("\nROI vs Driver Performance Score")
print("Correlation:", corr_roi)
print("P-value:", p_roi)

# 5.2 Salary vs Performance
subset_salary = df_r.dropna(subset=["estimatedSalaryUSD", "driverPerformanceScore"])
corr_sal, p_sal = pearsonr(subset_salary["estimatedSalaryUSD"], subset_salary["driverPerformanceScore"])

print("\nSalary vs Driver Performance Score")
print("Correlation:", corr_sal)
print("P-value:", p_sal)

# 5.3 Budget vs Constructor Points
subset_budget = df_r.dropna(subset=["costCapUSD", "constructorPoints"])
corr_budget, p_budget = pearsonr(subset_budget["costCapUSD"], subset_budget["constructorPoints"])

print("\nBudget vs Constructor Points")
print("Correlation:", corr_budget)
print("P-value:", p_budget)

# 5.4 Speed vs Performance
subset_speed = df_r.dropna(subset=["maxSpeed_kmh", "driverPerformanceScore"])
corr_speed, p_speed = pearsonr(subset_speed["maxSpeed_kmh"], subset_speed["driverPerformanceScore"])

print("\nTop Speed vs Driver Performance Score")
print("Correlation:", corr_speed)
print("P-value:", p_speed)

# ============================================================
# 6) SEGMENTATION PILOTES
# ============================================================
print("\n==============================")
print("SEGMENTATION PILOTES")
print("==============================")

segmentation = (
    df_r.groupby("driverName")[["driverPerformanceScore", "driverEfficiency", "driverROI"]]
    .mean()
    .dropna()
)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
segmentation["cluster"] = kmeans.fit_predict(segmentation)

print(segmentation.sort_values("driverPerformanceScore", ascending=False).head(20))

plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=segmentation,
    x="driverEfficiency",
    y="driverPerformanceScore",
    hue="cluster",
    palette="Set2"
)
plt.title("Segmentation Pilotes")
plt.show()

# ============================================================
# 7) VALIDATION HYPOTHÈSES
# ============================================================
print("\n==============================")
print("VALIDATION HYPOTHÈSES")
print("==============================")

print("\nHypothèse 1 : Plus un pilote est payé, plus il est performant ?")
print(f"Corr = {corr_sal:.3f}, p-value = {p_sal:.5f}")

print("\nHypothèse 2 : Plus le budget est élevé, meilleure est l’écurie ?")
print(f"Corr = {corr_budget:.3f}, p-value = {p_budget:.5f}")

print("\nHypothèse 3 : La vitesse max influence la performance pilote ?")
print(f"Corr = {corr_speed:.3f}, p-value = {p_speed:.5f}")

# ============================================================
# 8) ÉVOLUTION PAR SAISON
# ============================================================
print("\n==============================")
print("ÉVOLUTION PAR SAISON")
print("==============================")

season_perf = (
    df_r.groupby("season")[["driverPerformanceScore", "driverEfficiency", "raceGain"]]
    .mean()
)

print(season_perf)

season_perf.plot(figsize=(10, 5), marker="o")
plt.title("Évolution moyenne des KPI par saison")
plt.ylabel("Valeur moyenne")
plt.show()

# ============================================================
# 9) EFFICACITÉ ÉQUIPES
# ============================================================
print("\n==============================")
print("EFFICACITÉ DES ÉQUIPES")
print("==============================")

team_efficiency = (
    df_r.groupby("constructorName")[["constructorPoints", "budgetEfficiency", "pointsPerMillionUSD"]]
    .mean()
    .sort_values("constructorPoints", ascending=False)
)

print(team_efficiency)

# ============================================================
# 10) IMPACT DU DÉPART SUR LE RÉSULTAT
# ============================================================
print("\n==============================")
print("IMPACT DU DÉPART SUR LE RÉSULTAT")
print("==============================")

impact_depart = df_r[["gridPosition", "finishPosition"]].corr()
print(impact_depart)

plt.figure(figsize=(8, 5))
sns.regplot(data=df_r, x="gridPosition", y="finishPosition")
plt.title("Impact de la position de départ sur l’arrivée")
plt.xlabel("Grid Position")
plt.ylabel("Finish Position")
plt.show()

# ============================================================
# 11) ANALYSE DES GAINS DE POSITION
# ============================================================
print("\n==============================")
print("GAINS DE POSITION")
print("==============================")

print(df_r["raceGain"].describe())

plt.figure(figsize=(8, 5))
sns.histplot(df_r["raceGain"].dropna(), bins=20, kde=True)
plt.title("Distribution du Race Gain")
plt.xlabel("Race Gain")
plt.show()

# ============================================================
# 12) TOP DRIVERS / TEAMS
# ============================================================
print("\n==============================")
print("TOP DRIVERS / TEAMS")
print("==============================")

top_drivers = (
    df_r.groupby("driverName")["driverPerformanceScore"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

top_teams = (
    df_r.groupby("constructorName")["constructorPoints"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

print("\nTop Drivers:")
print(top_drivers)

print("\nTop Teams:")
print(top_teams)

```
